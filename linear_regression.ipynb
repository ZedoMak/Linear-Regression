```python
import numpy as np
from sklearn.ensemble._hist_gradient_boosting import predictor
```

```python
X= np.array([1.0, 1.5, 2.0, 2.5, 3.0])
y = np.array([250, 310, 400, 440, 520])

# compute the cost function
def compute_cost (X, y, w, b):
    n = len(X)
    predictions = w * X + b
    cost = (1 / (2 * n)) * np.sum((predictions - y) ** 2)
    return cost

# Compute Gradient and Update Parameters

def gradient_descent(X ,y, w_in, b_in, alpha, epochs):
    n = len(X)
    w = w_in
    b = b_in

    # track the cost history, this will help us to see if it'a actually decreasing

    cost_history = []

    for i in range(epochs):
        predictions = w * X + b  # calculate predictions
        dj_dw = (1 /n ) * np.sum((predictions - y) * X) # derivative of weight
        dj_db = (1 / n) * np.sum(predictions - y)  # derivative of bias

        w = w -alpha * dj_dw  # update weight
        b = b - alpha * dj_db # update bias

        current_cost = compute_cost(X, y, w, b)
        cost_history.append(current_cost)

        if i % 100 == 0:
            print(f"Epoch {i:4d}: Cost = {current_cost:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b, cost_history


initial_w = 0.0
initial_b = 0.0
learning_rate = 0.05
training_epochs = 1000


print("Starting Gradient Descent...")
print("-" * 40)
final_w, final_b, history = gradient_descent(X, y, initial_w, initial_b, learning_rate, training_epochs)
print("-" * 40)
print(f"Training complete! Final parameters: w = {final_w:.2f}, b = {final_b:.2f}")


#calculate predictions

```

```python
target_size = 2.2
predicted_price = final_w * target_size + final_b

print(f"Predicted price for a 2,200 sq ft house: ${predicted_price * 1000:,.2f}")
```

```python
import streamlit as st
import numpy as np

# --- THE ENGINE (Your NumPy Logic) ---
def predict_price(sqft, bedrooms, w, b):
    # Features: [sqft, bedrooms]
    x = np.array([sqft, bedrooms])
    return np.dot(w, x) + b

# Mock trained weights (after running your Gradient Descent)
# w1 (price per 1k sqft), w2 (price per bedroom)
trained_w = np.array([120.5, 25.0])
trained_b = 80.0

# --- THE INTERFACE (Streamlit) ---
st.title("🏡 House Price Predictor")
st.write("Deepen your understanding of ML by interacting with my Linear Regression model.")

st.divider()

# User Inputs
col1, col2 = st.columns(2)

with col1:
    sqft_input = st.number_input("Square Footage (in 1,000s)", min_value=0.5, max_value=10.0, value=1.5)

with col2:
    bedroom_input = st.slider("Number of Bedrooms", 1, 5, 2)

# Calculation
price = predict_price(sqft_input, bedroom_input, trained_w, trained_b)

# Visual Result
st.subheader(f"Estimated Value: :green[${price:,.2f}K]")

# Explanation for the user
st.info(f"""
**How this was calculated:**
The model used the formula:
$Price = ({trained_w[0]} \\times {sqft_input}) + ({trained_w[1]} \\times {bedroom_input}) + {trained_b}$
""")
```
